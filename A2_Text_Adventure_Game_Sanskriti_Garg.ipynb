<h2>Game Introduction</h2>

The game is based on **Season 5 of TV series F.R.I.E.N.D.S.** 

The game follows the secret dating between Monica and Chandler and how they did not want Joey, Rachel, Ross and Phoebe to know about them. Hence, they find different ways to avoid others getting to know about them. 

In the game set up, Monica is trying to reach the movie theatre where Chandler is waiting for her. But Joey, Rachel and Phoebe keep coming up to her and delaying her. The player needs to find ways to distract each of the friends and help Monica meet Chandler at the theatre and catch the movie on time

<hr style="height:.9px;border:none;color:#333;background-color:#333;" />
<h2>Game Rules</h2>

1.	The player will be presented with **3 situations for each friend**. 
2.	They need to choose the situation best suited to make the friend go away so that Monica can reach for her date with Chandler on time.
3.	They will get **2 chances** to choose the right situation. For each friend, 2 out of 3 given situations are the right ones. 
4. Out of the 2 correct situations for each friend, one must be further solved by the player but the other one is already solved and gives an easy pass to the user.
5.	For the situation that needs to be solved, player will have to follow the given instructions. Again, they will get **2 chances** to do that.
